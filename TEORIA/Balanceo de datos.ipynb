# Lectura de datos y recuento

```python
#Paquetes para lectura y recuento
from sklearn.datasets import load_breast_cancer
import pandas as pd

#Primero importamos los datos que vamos a leer
cancer_data = load_breast_cancer()

#Pasamos los datos a un DataFrame (datos en pandas)
data = pd.DataFrame(cancer_data['data'], columns = cancer_data['feature_names'])
target = pd.DataFrame(cancer_data['target'],columns = ['Tumor Type'])
Benigno = target.sum()
Maligno = target.shape[0]- Benigno
print('Tumores Benignos: ', Benigno, ' , Tumores Malignos: ', Maligno )
```

```python
data
```

# Sobremuestreo

Random Over-Sampler (ROS)
=> Duplicar entradas de la clase menos representada

SMOTE
=> Selecciona caracteristicas de las dos clases y calcula la media ponderada de esas características, agregando nuevas entradas son caracteristicas similares (pero no iguales) a las entradas de la clase menos representada.

=> SMOTE es más potente método de sobremuestreo

```python
#Importamos los paquetes
from imblearn.over_sampling import RandomOverSampler, SMOTE

#Inicialización de los métodos de sobremuestreo
#ROS
ros = RandomOverSampler()#random_state = 0
#SMOTE
smote = SMOTE()
```

```python
#ROS. Duplica muestras de la clase menos representadas
dataRos, targetRos = ros.fit_resample(data,target)
BenignoRos = targetRos.sum()
MalignoRos = targetRos.shape[0]- BenignoRos
print('Tumores Benignos: ', BenignoRos, ' , Tumores Malignos: ', MalignoRos )
dataRos
```

```python
#SMOTE. Genera nuevas muestras sintéticas
dataSmote, targetSmote = smote.fit_resample(data,target)
BenignoSmote = targetSmote.sum()
MalignoSmote = targetSmote.shape[0]- BenignoSmote
print('Tumores Benignos: ', BenignoSmote, ' , Tumores Malignos: ', MalignoSmote )
dataSmote
```

# Submuestreo

Random Under_Sampler (RUS) => Quitar de manera aleatoria muestras de la clase más representada

Perdidas cercanas (Near - Miss) = Quita muestras mas cercanas (con características más similares) a la clase minoritaria. Crea más diferencia entre las dos clases

```python
#Importamos los paquetes
from imblearn.under_sampling import RandomUnderSampler, NearMiss

#Inicialización de los métodos de submuestreo
#RUS
rus = RandomUnderSampler()#random_state = 0
#NearMiss
nm = NearMiss()
```

```python
#RUS. Elimina muestras de la clase más representada aleatoriamente
dataRus, targetRus = rus.fit_resample(data,target)
BenignoRus = targetRus.sum()
MalignoRus = targetRus.shape[0]- BenignoRus
print('Tumores Benignos: ', BenignoRus, ' , Tumores Malignos: ', MalignoRus )
dataRus
```

```python
#NearMiss. Elimina las muestras más cercanas de la clase más representada
dataNm, targetNm = nm.fit_resample(data,target)
BenignoNm = targetNm.sum()
MalignoNm = targetNm.shape[0]- BenignoNm
print('Tumores Benignos: ', BenignoNm, ' , Tumores Malignos: ', MalignoNm )
dataNm
```

# Sobremuestreo + Submuestreo

SMOTE + Uniones de Tomek

SMOTE =>  Crea muestras sintetícas

Uniones de Tomek => Elimina muestras cercanas

```python
#Importamos los paquetes
from imblearn.combine import SMOTETomek

#Inicialización del método combinado
#SMOTE Tomek
smoteT = SMOTETomek()#random_state = 0
```

```python
#Smote-Tomek. Sobremuestreo con Smote seguido de un submuestreo con Uniones de Tomek
datasmoteT, targetsmoteT = smoteT.fit_resample(data,target)
BenignosmoteT = targetsmoteT.sum()
MalignosmoteT = targetsmoteT.shape[0]- BenignosmoteT
print('Tumores Benignos: ', BenignosmoteT, ' , Tumores Malignos: ', MalignosmoteT )
datasmoteT
```
